```python
import pandas as pd
import matplotlib.pyplot as plt

#arquivo = pd.read_csv('Brasil_Escola_Eixo_Apoio_Tecnico_Financeiro.csv')
#display(arquivo)
```

```python
tabela = pd.read_csv('dados_educacao_github.csv')
display(tabela)
```

```python
tabela2 = pd.read_csv('Taxa_analfabetismo2.csv')
display(tabela2)
```

```python
tabela2.columns = tabela2.columns.str.strip()
print(tabela2.columns)
```

Fazendo a média para os gráficos


```python
analfabetismo = tabela2[['Regiões', 'Taxa de analfabetismo de pessoas acima de 15 anos - 2018', 'Taxa de analfabetismo de pessoas acima de 15 anos - 2020', 'Taxa de analfabetismo de pessoas acima de 15 anos - 2022']].groupby(['Regiões']).sum()

contagem2 = tabela2.groupby(['Regiões']).size()

analfabetismo['Média da taxa de analfabetismo - 2018'] = analfabetismo['Taxa de analfabetismo de pessoas acima de 15 anos - 2018'] / contagem2
analfabetismo['Média da taxa de analfabetismo - 2020'] = analfabetismo['Taxa de analfabetismo de pessoas acima de 15 anos - 2020'] / contagem2
analfabetismo['Média da taxa de analfabetismo - 2022'] = analfabetismo['Taxa de analfabetismo de pessoas acima de 15 anos - 2022'] / contagem2

analfabetismo = analfabetismo.reset_index()

display(analfabetismo)
```

```python

escolaridade = tabela[['regiões', 'matrículas da rede pública no ensino fundamental 2017', 'matrículas da rede privada no ensino fundamental 2017', 'matrículas da rede pública no ensino médio 2017', 'matrículas da rede privada no ensino médio 2017']].groupby(['regiões']).sum()

contagem = tabela.groupby(['regiões']).size()

escolaridade['Média Ensino Fundamental'] = escolaridade['matrículas da rede pública no ensino fundamental 2017'] / contagem
escolaridade['Média Ensino Fundamental Privado'] = escolaridade['matrículas da rede privada no ensino fundamental 2017'] / contagem
escolaridade['Média Ensino Médio'] = escolaridade['matrículas da rede pública no ensino médio 2017'] / contagem
escolaridade['Média Ensino Médio Privado'] = escolaridade['matrículas da rede privada no ensino médio 2017'] / contagem

escolaridade = escolaridade.reset_index()

display(escolaridade)
```

```python
escolaridade_total = tabela[['regiões', 'matrículas da rede pública no ensino fundamental 2017', 'matrículas da rede privada no ensino fundamental 2017', 'matrículas da rede pública no ensino médio 2017', 'matrículas da rede privada no ensino médio 2017']].groupby(['regiões']).sum()

contagem = tabela.groupby(['regiões']).size()

escolaridade_total['Média Ensino Fundamental'] = escolaridade_total['matrículas da rede pública no ensino fundamental 2017'] / contagem
escolaridade_total['Média Ensino Fundamental Privado'] = escolaridade_total['matrículas da rede privada no ensino fundamental 2017'] / contagem
escolaridade_total['Média Ensino Médio'] = escolaridade_total['matrículas da rede pública no ensino médio 2017'] / contagem
escolaridade_total['Média Ensino Médio Privado'] = escolaridade_total['matrículas da rede privada no ensino médio 2017'] / contagem

escolaridade_total = escolaridade_total.reset_index()

brasil = escolaridade_total[['matrículas da rede pública no ensino fundamental 2017', 'matrículas da rede privada no ensino fundamental 2017', 'matrículas da rede pública no ensino médio 2017', 'matrículas da rede privada no ensino médio 2017']].sum()

total_matriculas_publicas_fundamental = escolaridade_total['matrículas da rede pública no ensino fundamental 2017'].sum()
total_matriculas_privadas_fundamental = escolaridade_total['matrículas da rede privada no ensino fundamental 2017'].sum()
total_matriculas_publicas_medio = escolaridade_total['matrículas da rede pública no ensino médio 2017'].sum()
total_matriculas_privadas_medio = escolaridade_total['matrículas da rede privada no ensino médio 2017'].sum()

brasil['Média Ensino Fundamental'] = (escolaridade_total['Média Ensino Fundamental'] * escolaridade_total['matrículas da rede pública no ensino fundamental 2017']).sum() / total_matriculas_publicas_fundamental
brasil['Média Ensino Fundamental Privado'] = (escolaridade_total['Média Ensino Fundamental Privado'] * escolaridade_total['matrículas da rede privada no ensino fundamental 2017']).sum() / total_matriculas_privadas_fundamental
brasil['Média Ensino Médio'] = (escolaridade_total['Média Ensino Médio'] * escolaridade_total['matrículas da rede pública no ensino médio 2017']).sum() / total_matriculas_publicas_medio
brasil['Média Ensino Médio Privado'] = (escolaridade_total['Média Ensino Médio Privado'] * escolaridade_total['matrículas da rede privada no ensino médio 2017']).sum() / total_matriculas_privadas_medio

brasil['regiões'] = 'Brasil'

brasil_data = pd.DataFrame([brasil], columns=escolaridade_total.columns)

escolaridade_total = pd.concat([escolaridade_total, brasil_data], ignore_index=True)

escolaridade_total = brasil_data

display(escolaridade_total)
```

Gráfico Pizza*

Média Ensino Fundamental

```python
media_f = escolaridade_total[['Média Ensino Fundamental', 'Média Ensino Fundamental Privado']].values.flatten()

plt.figure(figsize=(8, 8))
plt.pie(
    media_f,
    labels=None, 
    autopct="%1.2f%%",
    textprops={'color': 'white', 'fontsize': 14,},
    startangle=90, 
    wedgeprops=dict(width=0.5, edgecolor='w'),
    pctdistance=0.73,
    colors=['#483D8B', '#008B8B']
)

plt.text(0, 0, 'Brasil', ha='center', color='gray', va='center', fontsize='18')

plt.legend(['Ensino Fundamental Público', 'Ensino Fundamental Privado'], title='Modalidades', loc='lower left', prop={'size': 12})

plt.suptitle('Média de Matrículas do Ensino Fundamental no Brasil', fontweight='bold', y=0.94)
plt.title('Destacando a diferença em relação a matrículas de escolas públicas e privadas', color='gray', fontsize=10)

plt.axis('equal')

plt.savefig("MédiaFundamental.png")

plt.show()
```
